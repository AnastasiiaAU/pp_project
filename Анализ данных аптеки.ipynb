- Критерии важности от 1 до 5, где 5 - важен, а 1 - не важен
- Критерии удовлетворенности от 1 до 5, где 5 - полностью удовлетворен, а 1 - полностью неудовлетворен
- Критерий частоты посещения от наиболее редкого до наиболее частого, 1-6
- Возраст: от наименьшей по возрасту группы к наибольшей от 1 до 4
- коды: женский пол - 1, мужской - 0
- Уровень дохода от наименьшего к наибольшему (1-6)
- Запомнил/не запомнил: 1/0

```python
#подготовка датафрейма
import pandas as pd
import numpy as np
data = pd.read_excel('ответы.xlsx')
def vaznost(string):
    if string == 'Очень важно': return 5
    elif string == 'Скорее важно': return 4
    elif string == 'Затрудняюсь ответить': return 3
    elif string == 'Скорее не важно': return 2
    elif string == 'Совершенно не важно': return 1

data['Важность_Удобство_расположения_аптеки'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Удобство расположения аптеки]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Удобство расположения аптеки]']

data['Важность_Квалификация_фармацевтов_аптеки'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Квалификация фармацевтов аптеки]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Квалификация фармацевтов аптеки]']

data['Важность_Дружелюбность_персонала_аптеки'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Дружелюбность персонала аптеки]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Дружелюбность персонала аптеки]']

data['Важность_Ассортимент_лекарств_в_аптеке'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Ассортимент лекарств в аптеке]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Ассортимент лекарств в аптеке]']

data['Важность_Ассортимент_других_товаров_в_аптеке'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Ассортимент других товаров в аптеке (вода, предметы личной гигиены и т.д.)]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Ассортимент других товаров в аптеке (вода, предметы личной гигиены и т.д.)]']

data['Важность_Уровень_цен'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Уровень цен]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Уровень цен]']

data['Важность_Скорость_обслуживания_фармацевта'] = data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Скорость обслуживания фармацевта]'].apply(vaznost)
del data['Оцените по шкале от 1 до 5 важность каждой из следующих характеристик при посещении аптеки: [Скорость обслуживания фармацевта]']

def recognize(x):
    x = x.split(', ')
    for i in x:
        if i == '36,6': return 1
    return 0

data['Запомнил_36_6'] = data['Какие бренды аптек вы можете вспомнить?'].apply(recognize)
del data['Какие бренды аптек вы можете вспомнить?']

def vaznost(string):
    if string == 'Полностью удовлетворен': return 5
    elif string == 'Скорее удовлетворен': return 4
    elif string == 'Затрудняюсь ответить': return 3
    elif string == 'Скорее не удовлетворен': return 2
    elif string == 'Совершенно не удовлетворен': return 1

data['Оценка_Удобство_расположения_аптеки'] = data['Оцените аптеку 36,6 по следующим параметрам: [Удобство расположения аптеки]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Удобство расположения аптеки]']

data['Оценка_Квалификация_фармацевтов_аптеки'] = data['Оцените аптеку 36,6 по следующим параметрам: [Квалификация фармацевтов аптеки]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Квалификация фармацевтов аптеки]']

data['Оценка_Дружелюбность_персонала_аптеки'] = data['Оцените аптеку 36,6 по следующим параметрам: [Дружелюбность персонала аптеки]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Дружелюбность персонала аптеки]']

data['Оценка_Ассортимент_лекарств_в_аптеке'] = data['Оцените аптеку 36,6 по следующим параметрам: [Ассортимент лекарств в аптеке]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Ассортимент лекарств в аптеке]']

data['Оценка_Уровень_цен'] = data['Оцените аптеку 36,6 по следующим параметрам: [Уровень цен]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Уровень цен]']

data['Оценка_Время_ожидания_в_очереди'] = data['Оцените аптеку 36,6 по следующим параметрам: [Время ожидания в очереди]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Время ожидания в очереди]']

data['Оценка_Чистота_помещения'] = data['Оцените аптеку 36,6 по следующим параметрам: [Чистота помещения аптеки]'].apply(vaznost)
del data['Оцените аптеку 36,6 по следующим параметрам: [Чистота помещения аптеки]']

def vaznost(string):
    if string == 'Очень важно': return 5
    elif string == 'Скорее важно': return 4
    elif string == 'Затрудняюсь ответить': return 3
    elif string == 'Скорее не важно': return 2
    elif string == 'Не важно': return 1
    
data['Важность_наличия_Лекарственные_препараты'] = data['Оцените насколько важно для Вас наличие в аптеке: [Лекарственные препараты]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Лекарственные препараты]']

data['Важность_наличия_Лекарственное_растительное_сырье'] = data['Оцените насколько важно для Вас наличие в аптеке: [Лекарственное растительное сырье в заводской упаковке (ромашки цветки, корень шалфея и т.п)]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Лекарственное растительное сырье в заводской упаковке (ромашки цветки, корень шалфея и т.п)]']

data['Важность_наличия_Дезинфицирующие_средства'] = data['Оцените насколько важно для Вас наличие в аптеке: [Дезинфицирующие средства]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Дезинфицирующие средства]']

data['Важность_наличия_Медицинские_изделия'] = data['Оцените насколько важно для Вас наличие в аптеке: [М﻿﻿едицинские изделия (разные инструменты, приборы и аппараты)]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [М﻿﻿едицинские изделия (разные инструменты, приборы и аппараты)]']

data['Важность_наличия_Предметы_личной_гигиены'] = data['Оцените насколько важно для Вас наличие в аптеке: [Предметы личной гигиены]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Предметы личной гигиены]']

data['Важность_наличия_Лечебное_и_детское_питание'] = data['Оцените насколько важно для Вас наличие в аптеке: [Лечебное и детское питание]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Лечебное и детское питание]']

data['Важность_наличия_Минеральная_вода'] = data['Оцените насколько важно для Вас наличие в аптеке: [Минеральная вода]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Минеральная вода]']

data['Важность_наличия_Оптика'] = data['Оцените насколько важно для Вас наличие в аптеке: [Оптика]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Оптика]']

data['Важность_наличия_Парфюмерная_и_косметическая_продукция'] = data['Оцените насколько важно для Вас наличие в аптеке: [Парфюмерная и косметическая продукция]'].apply(vaznost)
del data['Оцените насколько важно для Вас наличие в аптеке: [Парфюмерная и косметическая продукция]']

def poseshenie(x):
    if x == 'Несколько раз в месяц': return 4
    elif x == 'Раз в месяц': return 3
    elif x == 'Раз в несколько месяцев': return 2
    elif x == 'Раз в полгода и реже': return 1
    elif x == 'Раз в неделю': return 5
    elif x == 'Несколько раз в неделю': return 6

data['Частота_использования_аптек'] = data['Как часто вы пользуетесь услугами аптек?'].apply(poseshenie)
del data['Как часто вы пользуетесь услугами аптек?']

data['Частота_пользования_36_6'] = data['Как часто вы посещаете данные сети аптек? [36,6]'].apply(poseshenie)
del data['Как часто вы посещаете данные сети аптек? [36,6]']

data['Пол'] = data['Укажите ваш пол'].apply(lambda x: 1 if 'Женский' else 0)
del data['Укажите ваш пол']
def age(x):
    if x == 'До 18': return 1
    elif x == '18-29': return 2
    elif x == '30-49': return 3
    elif x == '50-69': return 4
data['Возраст'] = data['Укажите ваш возраст'].apply(age)
del data['Укажите ваш возраст']
data['Вероятность_рекомендации'] = data['Какова вероятность того, что вы порекомендуете перечисленные сети аптек другу или коллеге (1 – маловероятно, 10 – с большой вероятностью)? [36,6]']
del data['Какова вероятность того, что вы порекомендуете перечисленные сети аптек другу или коллеге (1 – маловероятно, 10 – с большой вероятностью)? [36,6]']
del data['Укажите ваш город проживания']
def dohod(x):
    if x == 'Денег не хватает даже на продукты, «едва сводим концы с концами»': return 1
    elif x == 'На продукты денег хватает, но покупка одежды вызывает затруднения': return 2
    elif x == 'Денег хватает на продукты и одежду, но покупка крупной бытовой техники является для нас затруднительной': return 3
    elif x == 'Можем без труда приобретать крупную бытовую технику, но покупка нового легкового автомобиля была бы затруднительной': return 4
    elif x == 'Хватает доходов на новый легковой автомобиль, однако покупка квартиры или дома (иной недвижимости) является для нас затруднительной': return 5
    elif x == 'Материальных затруднений не испытываем; при необходимости можем купить квартиру, дом (иную недвижимость)': return 6
data['Доход'] = data['Укажите ваш уровень дохода'].apply(dohod)
del data['Укажите ваш уровень дохода']
del data['Если бы вы могли что-то изменить или улучшить в какой-либо из сетей аптек или в аптеках в целом, что бы это было?']
```

```python
#гипотеза №1
import statsmodels.formula.api as smf
# построение регрессионной модели по избранным факторам
lm = smf.ols(formula='Запомнил_36_6 ~ Оценка_Время_ожидания_в_очереди + Оценка_Чистота_помещения + Оценка_Ассортимент_лекарств_в_аптеке + Оценка_Уровень_цен + Оценка_Дружелюбность_персонала_аптеки + Оценка_Квалификация_фармацевтов_аптеки + Оценка_Удобство_расположения_аптеки', data=data).fit()

# вывод коэффициентов
lm.params
lm.summary() # вывод саммари по модели
```

```python
#гипотеза №2
# построение регрессионной модели по избранным факторам
lm = smf.ols(formula='Вероятность_рекомендации ~ Частота_пользования_36_6 + Частота_использования_аптек', data=data).fit()

# вывод коэффициентов
lm.params
lm.summary() # вывод саммари по модели
```

```python
data['Удобство_расположения_аптеки'] = data['Важность_Удобство_расположения_аптеки'] / 5 * data['Оценка_Удобство_расположения_аптеки'] 
data['Квалификация_фармацевтов_аптеки'] = data['Важность_Квалификация_фармацевтов_аптеки'] / 5 * data['Оценка_Квалификация_фармацевтов_аптеки'] 
data['Дружелюбность_персонала_аптеки'] = data['Важность_Дружелюбность_персонала_аптеки'] / 5 * data['Оценка_Дружелюбность_персонала_аптеки'] 
data['Ассортимент_лекарств_в_аптеке'] = (data['Важность_Ассортимент_лекарств_в_аптеке'] + data['Важность_Ассортимент_других_товаров_в_аптеке'])/ 10 * data['Оценка_Ассортимент_лекарств_в_аптеке'] 
data['Уровень_цен'] = data['Важность_Уровень_цен'] / 5 * data['Оценка_Уровень_цен'] 
data['Скорость_обслуживания_фармацевта'] = data['Важность_Скорость_обслуживания_фармацевта'] / 5 * data['Оценка_Время_ожидания_в_очереди'] 
```

```python
#гипотеза №3
# построение регрессионной модели по избранным факторам
lm = smf.ols(formula='Вероятность_рекомендации ~ Удобство_расположения_аптеки + Квалификация_фармацевтов_аптеки + Дружелюбность_персонала_аптеки + Ассортимент_лекарств_в_аптеке + Уровень_цен + Скорость_обслуживания_фармацевта + Удобство_расположения_аптеки', data=data).fit()

# вывод коэффициентов
lm.params
lm.summary() # вывод саммари по модели
```
