```python
def fish(x):
    if x == 'Полностью удовлетворен': return 3
    elif x == 'Скорее удовлетворен': return 1.5
    elif x == 'Затрудняюсь ответить': return 0
    elif x == 'Скорее не удовлетворен': return -1.5
    elif x == 'Совершенно не удовлетворен': return -3
    else: return 0

sp = []
for item in fishben.columns:
    sp.append(item)
sp

for item in sp:
    fishben[item] = fishben[item].apply(fish)
fishben
```

```python
ind = ['Удобство расположения аптеки', 'Квалификация фармацевтов аптеки', 'Дружелюбность персонала аптеки', 
      'Ассортимент лекарств в аптеке', 'Уровень цен', 'Время ожидания в очереди', 'Чистота помещения аптеки']
col = ['Оценка ei', 'Горздрав', 'Ригла', 'Доктор Столетов', 'Столички', 'Неофарм']

final = []
for nazv in ind:
    data_mean = []
    for item in fishben.columns:
        if nazv in item: data_mean.append(round(fishben[item].mean(), 2))
    final.append(data_mean)
final
```

```python
def normalize_to_scale(data):
    # Flatten the list of lists to find the global min and max
    flat_list = [item for sublist in data for item in sublist]
    min_val = data[0][0]
    max_val = data[0][0]
    for i in data:
        for a in i:
            if a > max_val: max_val = a
            elif a < min_val: min_val = a
    
    # Function to scale an individual value
    def scale_value(value):
        return ((value - min_val) / (max_val - min_val)) * 6 - 3
    
    # Apply the scaling function to the data
    return [[scale_value(value) for value in sublist] for sublist in data]

new = normalize_to_scale(final)

for item in range(len(new)):
    for a in range(len(new[item])):
        new[item][a] = round(new[item][a], 2)
new
```
